```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/crimes.csv', sep=';', encoding='latin1')  # Abertura da base de dados
```

```python
data
```

```python
# Tratamento dos dados: Exclusão de colunas que não farão parte do somatório de crimes para a regressão linear
data_crimes = data.drop(columns=['fmun_cod','município','regiao','mes','mes_ano','total_roubos','total_furtos','registro_ocorrencias','fase'])
```

A coluna de total_roubos e total_furtos foi excluida, pois representavam um somatório das dos dados de outras colunas da base de dados.

```python
data_crimes # Visualização da base de dados após as exclusões das colunas
```

```python
# Agora temos na nossa base uma coluna chamada ano e as demais dos quantitativos de crimes.
# Queremos somar todas essas colunas para criar uma única.
crime_columns = data_crimes.columns.difference(['ano']) # Essa linha vai pegar os nomes de todas as colunas, com exceção da coluna ano.
```

```python
crime_columns # Nomes das colunas coletadas
```

```python
data_crimes['qntd_crimes'] = data[crime_columns].sum(axis=1) # Criação de uma nova coluna chamada qntd_crimes com o somatório de valores de todas as colunas de crimes
```

```python
data_crimes # Coluna qntd_crimes entrou como ultima coluna da base.
```

```python
data_ano = data_crimes.groupby('ano')['qntd_crimes'].sum().reset_index() # Agrupar valores da coluna de qnt_crimes por ano
```

```python
data_ano # Visualizando dataframe
```

```python
# Definindo os eixos x e y

x = data_ano[['ano']]
y = data_ano['qntd_crimes']

# Criar instância do modelo de regressão linear e ajustar modelo com dados x e y
model = LinearRegression()
model.fit(x, y)

# fazer previsões da relação x y
y_pred = model.predict(x)

# Plotar os dados e a linha de regressão
plt.figure(figsize=(10, 6))
plt.scatter(data_ano['ano'], y, color='blue', label='Dados reais')
plt.plot(data_ano['ano'], y_pred, color='red', linewidth=2, label='Linha de Regressão')

# Configurações do gráfico
plt.title('Regressão Linear da Criminalidade no Rio de Janeiro (2020-2022)')
plt.xlabel('Ano')
plt.ylabel('Número de Incidentes')
plt.legend()
plt.grid(True)

# Mostrar o gráfico
plt.show()
```
